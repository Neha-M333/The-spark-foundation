Task-3: Prediction using Decision Tree Algorithm

```python
import numpy as np
import pandas as pd
import sklearn.metrics as sm
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, classification_report
```

```python
data = pd.read_csv("Task 6 Decision tree.csv", index_col=0)
```

```python
target = data["Species"]
df = data.copy()
df = df.drop("Species", axis=1)
df.shape
```

```python
x = data.iloc[:, [0, 1, 2, 3]].values
le = LabelEncoder()
data["Species"] = le.fit_transform(data["Species"])
y = data["Species"].values
data.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print("Training split:", x_train.shape)
print("Testing split:", x_test.shape)
```

```python
dtree = DecisionTreeClassifier()
dtree.fit(x_train, y_train)
print("Deision Tree Classifier created successfully!")
```

```python
y_pred = dtree.predict(x_test)
print("Classification report:\n", classification_report(y_test, y_pred))
```

```python
print("Accuracy:", sm.accuracy_score(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
plt.figure(figsize = (20, 20))
tree = plot_tree(dtree, feature_names = df.columns, precision=2, rounded=True, filled=True, class_names=target.values)
```
