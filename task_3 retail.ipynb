IMPORT LIBRARIES

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("SampleSuperstore.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
data.info()
```

EDA

```python
plt.figure(figsize=(35,15))
ed=sns.countplot(x="State", data=data)
ed.set_xticklabels(ed.get_xticklabels(),rotation=90)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('State', fontsize=30) 
plt.ylabel('Count', fontsize=30)
plt.show()  
```

```python
plt.figure(figsize=(35,15))
state = data.groupby(['State'])['Sales'].mean().reset_index()
ed=sns.barplot(x = 'State', y='Sales', data=state)
ed.set_xticklabels(ed.get_xticklabels(),rotation=90)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('State', fontsize=30) 
plt.ylabel('Mean Sales', fontsize=30)
plt.show()
```

```python
plt.figure(figsize=(35,15))
state1 = data.groupby(['State'])['Profit'].mean().reset_index()
ed=sns.barplot(x = 'State', y='Profit', data=state1)
ed.set_xticklabels(ed.get_xticklabels(),rotation=90)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('State', fontsize=30) 
plt.ylabel('Mean Profit', fontsize=30)
plt.show()
```
